```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def encoder(l, dim, n):
    '''
    l: length of sequence
    dim: code dimension
    n: sinusoidal modulation param
    return: encoded sequence (l, dim)
    '''
    code = np.zeros((l, dim))
    for i in range(l):
        for j in np.arange(dim // 2):
            den = np.power(n, 2 * j / dim) 
            code[i, 2 * j    ] = np.sin(i / den)
            code[i, 2 * j + 1] = np.cos(i / den)
    return code

B = encoder(100, 64, 1000) # dim=512 and n=10000 in 'Attention is All You Need'
print(B.shape)
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(B.transpose())
fig.colorbar(cax)
ax.set_xlabel("sequence position")
ax.set_ylabel("code position")
plt.show()
```
